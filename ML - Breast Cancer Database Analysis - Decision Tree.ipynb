This project, like the one in "Evaluation 03 on MLP," focuses on the implementation of a Neural Network based on Decision Trees with the objective of diagnosing one of the most common and impactful cancers affecting women's health globally – Breast Cancer. The code utilizes data from a set of computationally calculated features derived from the digitization of examination images obtained through Fine Needle Aspirate (FNA) from patients.

In this code, we will be employing techniques such as preprocessing, imputation, and normalization to optimize the model's efficiency. The analysis includes performance metrics, visualizations of class distribution, and insights into the branches of the network (tree), aiming to contribute to advances in early detection and treatment of breast cancer.

```python
# Importing Library
import pandas as pd
import requests
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.impute import KNNImputer
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.exceptions import ConvergenceWarning
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_curve, auc, precision_recall_curve, average_precision_score
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import learning_curve, train_test_split

plt.close('all')
print('Results Obtained from the Program: ')

# Global Script Settings
UCI_DATASET_URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
DATA_FILENAME = "wdbc.data"
TARGET_COLUMN = "Diagnosis"
MAX_DEPTH = 5  # Maximum depth of the Decision Tree
RANDOM_STATE = 42
NUM_FOLDS = 5
NORMALIZATION_FLAG = 0  # 1 for ON, 0 for OFF
NUM_K = 5

### Suppress convergence warnings from MLPClassifier
warnings.filterwarnings("ignore", category=ConvergenceWarning)

### Function to download and load data
def download_and_load_data():
    # Download the breast cancer dataset from UCI
    url = UCI_DATASET_URL
    response = requests.get(url, verify=True)

    # Save the file to the local system
    with open(DATA_FILENAME, "wb") as file:
        file.write(response.content)

    # Read the data from the downloaded CSV file
    column_names = ["ID", "TARGET_COLUMN"] + [f"{feature}_{i}" for i in range(1, 4) for feature in ["radius", "texture", "perimeter", "area", "smoothness", "compactness", "concavity", "concave_points", "symmetry", "fractal_dimension"]]
    cancer_data = pd.read_csv(DATA_FILENAME, header=None, names=column_names)

    return cancer_data

### Function to display the initial table
def display_initial_table(cancer_data):
    print("\nInitial Table (First 5 and Last 5 values):")
    display(pd.concat([cancer_data.head(), pd.DataFrame(["..."] * (len(cancer_data) - 10)).set_index([0]), cancer_data.tail()]))

### Function to display the statistical summary of the data
def display_data_summary(cancer_data):
    print("\nStatistical Summary of the Data:")
    summary_table = cancer_data.describe(include='all').transpose().reset_index()
    summary_table = summary_table.rename(columns={'index': 'Feature'})
    display(summary_table)

### Function to visualize the distribution of classes
def visualize_class_distribution(cancer_data):
    plt.figure(figsize=(8, 6))
    diagnosis_counts = cancer_data['TARGET_COLUMN'].value_counts()
    diagnosis_counts.plot(kind='bar', color=['navy', 'red'])
    plt.title('Distribution of Diagnoses')
    plt.xlabel('Diagnosis')
    plt.ylabel('Count')
    plt.xticks(rotation=0, horizontalalignment="center")
    plt.show()

### Function to create a Correlation Matrix between the data
def plot_correlation_matrix(data):
    # Calculate the correlation matrix
    correlation_matrix = data.corr()

    # Visualize the correlation matrix using a Seaborn heatmap
    plt.figure(figsize=(12, 10))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
    plt.title('Correlation Matrix of the Data')
    plt.show()

### Function to preprocess and impute data
def preprocess_and_impute(data):
    data.replace('?', np.nan, inplace=True)
    missing_percentage = (data.isnull().sum() / len(data)) * 100
    missing_percentage = missing_percentage[missing_percentage > 0]
    if not missing_percentage.empty:
        print("Percentage of Missing Data by Column:")
        print(missing_percentage.sort_values(ascending=False))
    else:
        print("There is no missing data in the dataset.")

    data['TARGET_COLUMN'] = data['TARGET_COLUMN'].map({'M': 1, 'B': 0})

    data = data.where(pd.notna(data), None)
    attributes = data.drop('TARGET_COLUMN', axis=1)

    imputer_knn = KNNImputer(n_neighbors=NUM_K)
    attributes_imputed = imputer_knn.fit_transform(attributes)

    data_imputed = pd.DataFrame(attributes_imputed, columns=attributes.columns)
    data_imputed['TARGET_COLUMN'] = data['TARGET_COLUMN']

    scaler = StandardScaler()
    data_imputed.iloc[:, :-1] = scaler.fit_transform(data_imputed.iloc[:, :-1])

    return data_imputed

### Function that calculates the specificity metric.
def specificity_score(y_true, y_pred):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    specificity = tn / (tn + fp)
    return specificity

### Function that evaluates the model using various performance metrics.
def evaluate_model(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    specificity = specificity_score(y_true, y_pred)

    print(f"\n Performance Metrics for Evaluating the Decision Tree in Fold {NUM_K}")
    print(f"  - Accuracy: {accuracy:.4f}")
    print(f"  - Precision: {precision:.4f}")
    print(f"  - Recall: {recall:.4f}")
    print(f"  - F1-Score: {f1:.4f}")
    print(f"  - Specificity: {specificity_score(y_true, y_pred):.4f}")

    confusion = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(confusion, annot=True, fmt="d", cmap="Blues", xticklabels=['B', 'M'], yticklabels=['B', 'M'])
    plt.title(f"Confusion Matrix")
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()

### Function to Train a Decision Tree
def train_decision_tree_model(X, y, num_folds=NUM_FOLDS):
    processed_data = preprocess_and_impute(pd.concat([X, y], axis=1))
    X_processed = processed_data.drop(['TARGET_COLUMN'], axis=1)
    y_processed = processed_data['TARGET_COLUMN'].apply(lambda x: 1 if x == 'M' else 0)

    if NORMALIZATION_FLAG == 1:
        X_processed_scaled = StandardScaler().fit_transform(X_processed)
        print("\nNOTE: Normalized Database: YES")
    else:
        print("\nNOTE: Normalized Database: NO")

    accuracy_scores = []
    predicted_labels_folds = []
    y_test_folds = []

    model = DecisionTreeClassifier(max_depth=MAX_DEPTH, random_state=RANDOM_STATE)

    kfold = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=42)

    print("\nDecision Tree Training Analysis")

    for fold, (train_index, test_index) in enumerate(kfold.split(X, y), 1):

        if NORMALIZATION_FLAG == 1:
            X_train, X_test = X_processed_scaled[train_index], X_processed_scaled[test_index]
            y_train, y_test = y_processed.iloc[train_index], y_processed.iloc[test_index]
        else:
            X_train, X_test = X.iloc[train_index], X.iloc[test_index]
            y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        model.fit(X_train, y_train)

        predicted_labels = model.predict(X_test)
        predicted_labels_folds.append(predicted_labels)
        y_test_folds.append(y_test)

        accuracy = accuracy_score(y_test, predicted_labels)
        accuracy_scores.append(accuracy)

        print(f"Fold {fold} - Accuracy: {accuracy:.4f}")

    mean_accuracy = np.mean(accuracy_scores)
    print(f"\nMean Accuracy: {mean_accuracy:.4f}")

    evaluate_model(y_test, predicted_labels)

    if NORMALIZATION_FLAG == 1:
        model = DecisionTreeClassifier(max_depth=MAX_DEPTH, random_state=RANDOM_STATE)
        cross_val_scores = cross_val_score(model, X_processed_scaled, y_processed, cv=num_folds)
        print(f"\nCross-Validation Accuracy Scores: {cross_val_scores}")
        print(f"Mean Cross-Validation Accuracy: {cross_val_scores.mean():.4f}")
        model.fit(X_processed_scaled, y_processed)
    else:
        model = DecisionTreeClassifier(max_depth=MAX_DEPTH, random_state=RANDOM_STATE)
        model.fit(X_train, y_train)

    # Visualization of the Decision Tree
    plt.figure(figsize=(20, 10))
    plot_tree(model, feature_names=X_processed.columns, class_names=['B', 'M'], filled=True, rounded=True)
    plt.title(f"Decision Tree", fontsize=20)
    plt.show()

    # Display of Feature Importance
    feature_importance = model.feature_importances_
    feature_names = X_processed.columns
    feature_importance_df = pd.DataFrame({"Feature": feature_names, "Importance": feature_importance})
    feature_importance_df = feature_importance_df.sort_values(by="Importance", ascending=False)

    # Bar Chart for Feature Importance
    plt.figure(figsize=(10, 6))
    sns.barplot(x="Importance", y="Feature", data=feature_importance_df, palette="viridis")
    plt.title('Feature Importance')
    plt.show()

    print("\nFeature Importance:")
    display(feature_importance_df)

    return model


def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=-1, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure(figsize=(10, 6))
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt

def plot_roc_curve(y_test, y_pred_prob):
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    plt.show()

def plot_precision_recall_curve(y_test, y_pred_prob):
    precision, recall, thresholds = precision_recall_curve(y_test, y_pred_prob)
    average_precision = average_precision_score(y_test, y_pred_prob)

    plt.figure(figsize=(8, 6))
    plt.step(recall, precision, color='b', alpha=0.2, where='post')
    plt.fill_between(recall, precision, step='post', alpha=0.2, color='b')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])
    plt.title('Precision-Recall Curve: AP={0:0.2f}'.format(average_precision))
    plt.show()


'''Initialization of the Main Code calling the Previous Function Blocks'''
if __name__ == "__main__":

    # Initializing the raw data analysis of the chosen database
    cancer_data = download_and_load_data()
    display_initial_table(cancer_data)
    plot_correlation_matrix(processed_data)
    display_data_summary(cancer_data)
    visualize_class_distribution(cancer_data)

    # Preprocessing and imputation of the data
    processed_data = preprocess_and_impute(cancer_data)
    X_processed = processed_data.drop(['ID', 'TARGET_COLUMN'], axis=1)
    y_processed = processed_data['TARGET_COLUMN']

    # Create training and testing data
    X_train, X_test, y_train, y_test = train_test_split(X_processed, y_processed, test_size=0.2, random_state=RANDOM_STATE)

    # Train the model
    decision_tree_model = train_decision_tree_model(X_train, y_train)

    # Obtain predicted probabilities for the test set
    y_pred_prob = decision_tree_model.predict_proba(X_test)[:, 1]

    # Additional visualizations for the model
    plot_learning_curve(decision_tree_model, "Learning Curve", X_train, y_train, cv=5)
    plot_roc_curve(y_test, y_pred_prob)
    plot_precision_recall_curve(y_test, y_pred_prob)
```

CONSIDERATIONS:

Before anything else, let’s revisit the parameters set in the model’s input to better evaluate the presented behavior:

- Maximum Depth of the Decision Tree (MAX_DEPTH): 5
- Random State (RANDOM_STATE): 42
- Number of Folds in Cross-Validation (NUM_FOLDS): 5
- Normalization Flag (NORMALIZATION_FLAG): Disabled (0)
- Number of Neighbors for KNNImputer (NUM_K): 5

It is also important to remember that, in the original model presented, we will not be normalizing the input data because this can lead to overfitting, as occurred when we implemented a learning machine based on MLP, and since we do not want to lose the ability to generalize, we will not be activating it. In the model's output, we obtained the following results (beyond the visual ones):

- Accuracy: 0.9121
- Precision: 0.9062
- Recall: 0.8529
- F1-Score: 0.8788
- Specificity: 0.9474

The average accuracy of around 94% suggests that the model is performing well in classifying the dataset. When we compare this model with the previously implemented MLP model (also using 5 folds, 2 hidden layers, and normalization disabled), we obtained an average accuracy of 95%. This could be attributed to the nature of both methods, as MLP is more suited for more complex problems, and in the discussed work, we also conducted more extensive preprocessing (applying different tools than those used here).

We can also evaluate which attributes are most impactful for breast cancer cases using the model, with "concave_points_1" being the most significant (70%). This also provides us with powerful average analysis to identify which examination results need to be highlighted for the physician's attention.

Further evaluations can be conducted based on the analysis of the additional graphs included in the model, such as the ROC curve, which presented an AUC of 96%, indicating that the model has, on average, an excellent ability to distinguish between positive and negative classes.

If we analyze the Precision-Recall Curve, we can see that the AP was 93%, which indicates very good performance in the trade-off between precision and recall. An AP of 93% suggests that, on average, the model has high precision at different levels of recall. This means that when the model makes a positive prediction, there is a high probability it is correct.

An important note that guided much of our implementation was the maximum optimization of the decision-making capacity for benign cases, as, when it comes to breast cancer, this is the most sensitive data (since misclassifying a benign prediction causes much more harm than misclassifying a malignant prediction).
