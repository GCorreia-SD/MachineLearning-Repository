```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.impute import KNNImputer
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.exceptions import ConvergenceWarning
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_curve, auc, precision_recall_curve, average_precision_score
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import learning_curve

plt.close('all')
print('Resultados Obtidos no Programa: ')

# Configurações Globais do Script
UCI_DATASET_URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
DATA_FILENAME = "wdbc.data"
TARGET_COLUMN = "Diagnosis"
MAX_DEPTH = 5  # Profundidade máxima da Árvore de Decisão
RANDOM_STATE = 42
NUM_FOLDS = 5
NORMALIZATION_FLAG = 0  # 1 for ON, 0 for OFF
NUM_K = 5

### Suprimir os avisos de convergência do MLPClassifier
warnings.filterwarnings("ignore", category=ConvergenceWarning)

### Função para baixar e carregar dados
def download_and_load_data():
    # Download do conjunto de dados de câncer de mama da UCI
    url = UCI_DATASET_URL
    response = requests.get(url, verify=True)

    # Salvar o arquivo no sistema local
    with open(DATA_FILENAME, "wb") as file:
        file.write(response.content)

    # Ler os dados do arquivo CSV baixado
    column_names = ["ID", "TARGET_COLUMN"] + [f"{feature}_{i}" for i in range(1, 4) for feature in ["radius", "texture", "perimeter", "area", "smoothness", "compactness", "concavity", "concave_points", "symmetry", "fractal_dimension"]]
    cancer_data = pd.read_csv(DATA_FILENAME, header=None, names=column_names)

    return cancer_data

### Função para exibir a tabela inicial
def display_initial_table(cancer_data):
    print("\nTabela Inicial (Primeiros 5 e Últimos 5 valores):")
    display(pd.concat([cancer_data.head(), pd.DataFrame(["..."] * (len(cancer_data) - 10)).set_index([0]), cancer_data.tail()]))

### Função para exibir o resumo estatístico dos dados
def display_data_summary(cancer_data):
    print("\nResumo Estatístico dos Dados:")
    summary_table = cancer_data.describe(include='all').transpose().reset_index()
    summary_table = summary_table.rename(columns={'index': 'Feature'})
    display(summary_table)

### Função para visualizar a distribuição das classes
def visualize_class_distribution(cancer_data):
    plt.figure(figsize=(8, 6))
    diagnosis_counts = cancer_data['TARGET_COLUMN'].value_counts()
    diagnosis_counts.plot(kind='bar', color=['navy', 'red'])
    plt.title('Distribuição dos Diagnósticos')
    plt.xlabel('Diagnóstico')
    plt.ylabel('Contagem')
    plt.xticks(rotation=0, horizontalalignment="center")
    plt.show()

### Função para criar uma Matriz de Correlação entre os dados
def plot_correlation_matrix(data):
    # Calcular a matriz de correlação
    correlation_matrix = data.corr()

    # Visualizar a matriz de correlação usando um heatmap do Seaborn
    plt.figure(figsize=(12, 10))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
    plt.title('Matriz de Correlação dos Dados')
    plt.show()

### Função para pré-processar e imputar dados
def preprocess_and_impute(data):
    data.replace('?', np.nan, inplace=True)
    missing_percentage = (data.isnull().sum() / len(data)) * 100
    missing_percentage = missing_percentage[missing_percentage > 0]
    if not missing_percentage.empty:
        print("Porcentagem de Dados Faltantes por Coluna:")
        print(missing_percentage.sort_values(ascending=False))
    else:
        print("Não há dados faltantes no conjunto de dados.")

    data['TARGET_COLUMN'] = data['TARGET_COLUMN'].map({'M': 1, 'B': 0})

    data = data.where(pd.notna(data), None)
    attributes = data.drop('TARGET_COLUMN', axis=1)

    imputer_knn = KNNImputer(n_neighbors=NUM_K)
    attributes_imputed = imputer_knn.fit_transform(attributes)

    data_imputed = pd.DataFrame(attributes_imputed, columns=attributes.columns)
    data_imputed['TARGET_COLUMN'] = data['TARGET_COLUMN']

    scaler = StandardScaler()
    data_imputed.iloc[:, :-1] = scaler.fit_transform(data_imputed.iloc[:, :-1])

    return data_imputed

### Função que Calcula a métrica de especificidade.
def specificity_score(y_true, y_pred):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    specificity = tn / (tn + fp)
    return specificity

### Função que Avalia o modelo usando várias métricas de desempenho.
def evaluate_model(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    specificity = specificity_score(y_true, y_pred)

    print(f"\n Métricas de Desempenho para Avaliação da Árvore de Decisão no Fold {NUM_K}")
    print(f"  - Acurácia: {accuracy:.4f}")
    print(f"  - Precisão: {precision:.4f}")
    print(f"  - Recall: {recall:.4f}")
    print(f"  - F1-Score: {f1:.4f}")
    print(f"  - Especificidade: {specificity_score(y_true, y_pred):.4f}")

    confusion = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(confusion, annot=True, fmt="d", cmap="Blues", xticklabels=['B', 'M'], yticklabels=['B', 'M'])
    plt.title(f"Matriz de Confusão")
    plt.xlabel('Rótulos Previstos')
    plt.ylabel('Rótulos Verdadeiros')
    plt.show()

### Função para Treinar uma Árvore de Decisão
def train_decision_tree_model(X, y, num_folds=NUM_FOLDS):
    processed_data = preprocess_and_impute(pd.concat([X, y], axis=1))
    X_processed = processed_data.drop(['TARGET_COLUMN'], axis=1)
    y_processed = processed_data['TARGET_COLUMN'].apply(lambda x: 1 if x == 'M' else 0)

    if NORMALIZATION_FLAG == 1:
        X_processed_scaled = StandardScaler().fit_transform(X_processed)
        print("\nOBSERVAÇÃO: Base de Dados NORMALIZADOS: SIM")
    else:
        print("\nOBSERVAÇÃO: Base de Dados NORMALIZADOS: NÃO")

    accuracy_scores = []
    predicted_labels_folds = []
    y_test_folds = []

    model = DecisionTreeClassifier(max_depth=MAX_DEPTH, random_state=RANDOM_STATE)

    kfold = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=42)

    print("\nAnálise do Treinamento da Árvore de Decisão")

    for fold, (train_index, test_index) in enumerate(kfold.split(X, y), 1):

        if NORMALIZATION_FLAG == 1:
            X_train, X_test = X_processed_scaled[train_index], X_processed_scaled[test_index]
            y_train, y_test = y_processed.iloc[train_index], y_processed.iloc[test_index]
        else:
            X_train, X_test = X.iloc[train_index], X.iloc[test_index]
            y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        model.fit(X_train, y_train)

        predicted_labels = model.predict(X_test)
        predicted_labels_folds.append(predicted_labels)
        y_test_folds.append(y_test)

        accuracy = accuracy_score(y_test, predicted_labels)
        accuracy_scores.append(accuracy)

        print(f"Fold {fold} - Acurácia: {accuracy:.4f}")

    mean_accuracy = np.mean(accuracy_scores)
    print(f"\nAcurácia Média: {mean_accuracy:.4f}")

    evaluate_model(y_test, predicted_labels)

    if NORMALIZATION_FLAG == 1:
        model = DecisionTreeClassifier(max_depth=MAX_DEPTH, random_state=RANDOM_STATE)
        cross_val_scores = cross_val_score(model, X_processed_scaled, y_processed, cv=num_folds)
        print(f"\nPontuações de Acurácia na Validação Cruzada: {cross_val_scores}")
        print(f"Acurácia Média na Validação Cruzada: {cross_val_scores.mean():.4f}")
        model.fit(X_processed_scaled, y_processed)
    else:
        model = DecisionTreeClassifier(max_depth=MAX_DEPTH, random_state=RANDOM_STATE)
        model.fit(X_train, y_train)

    # Visualização da Árvore de Decisão
    plt.figure(figsize=(20, 10))
    plot_tree(model, feature_names=X_processed.columns, class_names=['B', 'M'], filled=True, rounded=True)
    plt.show()

    # Exibição da Importância das Características
    feature_importance = model.feature_importances_
    feature_names = X_processed.columns
    feature_importance_df = pd.DataFrame({"Feature": feature_names, "Importance": feature_importance})
    feature_importance_df = feature_importance_df.sort_values(by="Importance", ascending=False)

    # Gráfico de Barras para Importância das Características
    plt.figure(figsize=(10, 6))
    sns.barplot(x="Importance", y="Feature", data=feature_importance_df, palette="viridis")
    plt.title('Importância das Características')
    plt.show()

    print("\nImportância das Características:")
    display(feature_importance_df)

    return model


def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=-1, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure(figsize=(10, 6))
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt

def plot_roc_curve(y_test, y_pred_prob):
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    plt.show()

def plot_precision_recall_curve(y_test, y_pred_prob):
    precision, recall, thresholds = precision_recall_curve(y_test, y_pred_prob)
    average_precision = average_precision_score(y_test, y_pred_prob)

    plt.figure(figsize=(8, 6))
    plt.step(recall, precision, color='b', alpha=0.2, where='post')
    plt.fill_between(recall, precision, step='post', alpha=0.2, color='b')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])
    plt.title('Precision-Recall Curve: AP={0:0.2f}'.format(average_precision))
    plt.show()

'''Inicialização do Código Principal chamando os Blocos de Função Anteriores'''
if __name__ == "__main__":

    # Inicializando a análise bruta dos dados do banco de dados escolhido
    cancer_data = download_and_load_data()
    display_initial_table(cancer_data)
    plot_correlation_matrix(processed_data)
    display_data_summary(cancer_data)
    visualize_class_distribution(cancer_data)

    # Pré-processamento e imputação dos dados
    processed_data = preprocess_and_impute(cancer_data)
    X_processed = processed_data.drop(['ID', 'TARGET_COLUMN'], axis=1)
    y_processed = processed_data['TARGET_COLUMN']

    # Criar dados de treino e teste
    X_train, X_test, y_train, y_test = train_test_split(X_processed, y_processed, test_size=0.2, random_state=RANDOM_STATE)

    # Treinar o modelo
    decision_tree_model = train_decision_tree_model(X_train, y_train)

    # Obter probabilidades previstas para o conjunto de teste
    y_pred_prob = decision_tree_model.predict_proba(X_test)[:, 1]

    # Visualizações adicionais pra o modelo
    plot_learning_curve(decision_tree_model, "Learning Curve", X_train, y_train, cv=5)
    plot_roc_curve(y_test, y_pred_prob)
    plot_precision_recall_curve(y_test, y_pred_prob)
```

Antes de mais nada, vamos retornar os parâmetros colocados na entrada do modelo para avaliarmos melhor o comportamento apresentado: 

    - Profundidade máxima da Árvore de Decisão (MAX_DEPTH): 5
    - Estado aleatório (RANDOM_STATE): 42
    - Número de folds na validação cruzada (NUM_FOLDS): 5
    - Flag de normalização (NORMALIZATION_FLAG): Desativada (0)
    - Número de vizinhos para KNNImputer (NUM_K): 5

É também importante lembrar que, no modelo apresentado original, não estaremos normalizando os dados de entrada porque isso gera, como ocorreu quando implementamos uma máquina de aprendizado a partir do MLP, em OverFitting (e, como não queremos perder a capacidade de generalização, não estaremos ativando). Na saída do modelo, obtivemos os resultados (para além dos visuais): 

    - Acurácia: 0.9121
    - Precisão: 0.9062
    - Recall: 0.8529
    - F1-Score: 0.8788
    - Especificidade: 0.9474

A acurácia média em torno de 94% sugere que o modelo está performando bem na classificação dos dados da base. Se compararmos esse modelo com o modelo anteriormente implementado em MLP (estando também com 5Folds, 2 Camadas Ocultas e Normalização desativada), obtivemos um acurácia média de 95%. Isso pode se dar pela natureza de ambos os métodos, já que MLP é mais direcionado a problemas mais complexos e, no trabalho em questão comentado, também fizemos um trabalho de pré-processamento maior (aplicando outras ferramentas, diferentemente deste). 

Também podemos, com o modelo, avaliar quais atributos são mais impactantes para o caso de câncer de mama, o "concave_points_1" (70%). Isso, inclusive, nos fornece uma análise média poderosa para perceber qual dos resultados do exame mais precisa estar em atenção pelo médico. 

Outras avaliações podem ser realizadas a partir da análise dos gráficos adicionais inseridos no modelo, como a Curva ROC que apresentou uma AUC de 96%, o que significa que o modelo tem, em média, uma excelente capacidade de distinguir entre as classes positivas e negativas.

Se analisarmos a curva de precisão e recall (Precision Recall Curve), podemos perceber que a AP foi de 93%, o que indica um desempenho muito bom na trade-off entre precisão e recall. Uma AP de 93% sugere que, em média, o modelo tem uma alta precisão em diferentes níveis de recall. Isso significa que, quando o modelo faz uma previsão positiva, há uma alta probabilidade de estar correta.

Um adendo que é importante e que norteou grande parte da nossa implementação foi na máxima otimização de capacidade de decisão dos casos Benignos, uma vez que, tratando-se de câncer de mama, é o dado mais sensível (já que errar para Benigno na previsão gera muito mais danos do que errar a previsão para Maligno).
