```python
# Importing Library
import pandas as pd
import requests
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
import numpy as np
import tqdm
from sklearn.neural_network import MLPClassifier
from sklearn.impute import KNNImputer
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.exceptions import ConvergenceWarning
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import learning_curve
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
from tqdm import tqdm

plt.close('all')
print('Results Obtained in the Program: ')

# Global Settings of the Script (We Can Change This to Analyze Specific Behavior) THESE PARAMETERS CAN BE CHANGED TO ANALYZE DIFFERENT RESULTS
UCI_DATASET_URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
DATA_FILENAME = "wdbc.data"
TARGET_COLUMN = "Diagnosis"
NUM_HIDDEN_LAYERS = (8,)
MAX_ITER = 100
RANDOM_STATE = 42
NUM_FOLDS = 5
NORMALIZATION_FLAG = 0  # 1 for ON, 0 for OFF
NUM_K = 5

# Suppress convergence warnings from the MLPClassifier
warnings.filterwarnings("ignore", category=ConvergenceWarning)

def download_and_load_data():
    # Download the UCI breast cancer dataset
    url = UCI_DATASET_URL
    response = requests.get(url, verify=True)

    # Save the file to the local system
    with open(DATA_FILENAME, "wb") as file:
        file.write(response.content)

    # Read the data from the downloaded CSV file
    column_names = ["ID", "TARGET_COLUMN"] + [f"{feature}_{i}" for i in range(1, 4) for feature in ["radius", "texture", "perimeter", "area", "smoothness", "compactness", "concavity", "concave_points", "symmetry", "fractal_dimension"]]
    cancer_data = pd.read_csv("DATA_FILENAME", header=None, names=column_names)

    return cancer_data

def display_initial_table(cancer_data):
    # Display the initial table
    print("\nInitial Table (First 5 and Last 5 values):")
    display(pd.concat([cancer_data.head(), pd.DataFrame(["..."] * (len(cancer_data) - 10)).set_index([0]), cancer_data.tail()]))

def display_data_summary(cancer_data):
    # Statistical Summary of the Data
    print("\nStatistical Summary of the Data:")
    summary_table = cancer_data.describe(include='all').transpose().reset_index()
    summary_table = summary_table.rename(columns={'index': 'Feature'})
    display(summary_table)

def visualize_class_distribution(cancer_data):
    # Visualization of Class Distribution
    plt.figure(figsize=(8, 6))
    diagnosis_counts = cancer_data['TARGET_COLUMN'].value_counts()
    diagnosis_counts.plot(kind='bar', color=['navy', 'red'])
    plt.title('Diagnosis Distribution')
    plt.xlabel('Diagnosis')
    plt.ylabel('Count')
    plt.xticks(rotation=0, horizontalalignment="center")  # Adjust the text orientation on the X-axis
    plt.show()

def visualize_column_distribution(cancer_data):
    # Get number of columns
    num_cols = 3

    # Calculate total number of subplots needed
    num_features = len(cancer_data.columns[2:])
    num_rows = (num_features + num_cols - 1) // num_cols

    # Create the grid of subplots
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))

    # Adjust spacing between subplots
    plt.subplots_adjust(wspace=0.4, hspace=0.6)

    # Iterate over columns and plot graphs in subplots
    for i, column in enumerate(cancer_data.columns[2:]):
        row = i // num_cols
        col = i % num_cols

        # Count the number of categorical values
        count_values = cancer_data[column].value_counts()
        count_values.plot(kind='bar', color=['navy', 'red', 'gray'], ax=axes[row, col])

        axes[row, col].set_title(f'Value Count in {column}')
        axes[row, col].set_xlabel(column)
        axes[row, col].set_ylabel('Quantity')
        axes[row, col].set_xticklabels(count_values.index, rotation=0)

    # Remove empty subplots
    for i in range(num_features, num_rows * num_cols):
        fig.delaxes(axes.flatten()[i])

    # Display the figure
    plt.show()


def preprocess_and_impute(data):
    # Convert '?' to NaN
    data.replace('?', np.nan, inplace=True)

    # Check the percentage of missing data
    missing_percentage = (data.isnull().sum() / len(data)) * 100
    missing_percentage = missing_percentage[missing_percentage > 0]  # Only columns with missing data
    if not missing_percentage.empty:
        print("Percentage of Missing Data by Column:")
        print(missing_percentage.sort_values(ascending=False))
    else:
        print("There is no missing data in the dataset.")

    # Convert the columns to numeric values (0 and 1)
    data['TARGET_COLUMN'] = data['TARGET_COLUMN'].map({'M': 1, 'B': 0})

    # Replace NAType with None
    data = data.where(pd.notna(data), None)

    # Extract attribute columns (excluding the class column)
    attributes = data.drop('TARGET_COLUMN', axis=1)

    # Create the KNN imputer
    imputer_knn = KNNImputer(n_neighbors=NUM_K)

    # Fill missing values in attributes with the KNN imputer
    attributes_imputed = imputer_knn.fit_transform(attributes)

    # Create a new DataFrame with the filled values
    data_imputed = pd.DataFrame(attributes_imputed, columns=attributes.columns)

    # Add the class column to the new DataFrame
    data_imputed['TARGET_COLUMN'] = data['TARGET_COLUMN']

    # Normalize the data
    scaler = StandardScaler()
    data_imputed.iloc[:, :-1] = scaler.fit_transform(data_imputed.iloc[:, :-1])

    return data_imputed

def train_mlp_model(X, y, num_folds=NUM_FOLDS):
    # Preprocessing and imputation of data
    processed_data = preprocess_and_impute(pd.concat([X, y], axis=1))

    # It is not necessary to exclude 'ID' and 'Diagnosis' here, as the preprocess_and_impute function already includes 'Diagnosis'
    X_processed = processed_data.drop(['TARGET_COLUMN'], axis=1)
    y_processed = processed_data['TARGET_COLUMN'].apply(lambda x: 1 if x == 'M' else 0)

    if NORMALIZATION_FLAG==1:
        # Normalize the data (When this function is executed, all results reach their maximum values)
        scaler = StandardScaler()
        X_processed_scaled = scaler.fit_transform(X_processed)
        print("\nNOTE: Database NORMALIZED: YES")
    else:
        print("\nNOTE: Database NORMALIZED: NO")

    # Initialize lists to store performance metrics
    accuracy_scores = []
    precision_scores = []
    recall_scores = []
    f1_scores = []
    loss_curve = []

    # Initialize lists to store model parameters
    model_parameters = []

    # Include the trained model and its parameters in a list
    models_and_params = []

    # Create the StratifiedKFold object to split the data into folds
    kfold = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=42)

    print("\nAnalysis of MLP Network Training")

    # Iterate over the folds
    for fold, (train_index, test_index) in enumerate(kfold.split(X, y), 1):

        if NORMALIZATION_FLAG==1:
            # Split the data into training and testing (when there is normalization)
            X_train, X_test = X_processed_scaled[train_index], X_processed_scaled[test_index]
            y_train, y_test = y_processed.iloc[train_index], y_processed.iloc[test_index]
        else:
            # Split the data into training and testing (when there is no normalization)
            X_train, X_test = X.iloc[train_index], X.iloc[test_index]
            y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        # Create the MLP model
        model = MLPClassifier(hidden_layer_sizes=NUM_HIDDEN_LAYERS, max_iter=MAX_ITER, random_state=42)

        # Train the model
        model.fit(X_train, y_train)

        # Store the loss curve
        loss_curve.append(model.loss_curve_)

        # Store the model parameters
        model_parameters.append(model.get_params())

        # Evaluate the model's performance on the test data
        predicted_labels = model.predict(X_test)

        # Calculate performance metrics for the current fold
        accuracy = accuracy_score(y_test, predicted_labels)
        precision = precision_score(y_test, predicted_labels, average='weighted')
        recall = recall_score(y_test, predicted_labels, average='weighted')
        f1 = f1_score(y_test, predicted_labels, average='weighted')

        # Store the performance metrics
        accuracy_scores.append(accuracy)
        precision_scores.append(precision)
        recall_scores.append(recall)
        f1_scores.append(f1)

        # Display performance metrics for the current fold
        print(f"Fold {fold} - Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1-score: {f1:.4f}")

    # Calculate the means of the performance metrics
    mean_accuracy = np.mean(accuracy_scores)
    mean_precision = np.mean(precision_scores)
    mean_recall = np.mean(recall_scores)
    mean_f1 = np.mean(f1_scores)

    # Display the means of the performance metrics
    print(f"\nMean Accuracy: {mean_accuracy:.4f}")
    print(f"Mean Precision: {mean_precision:.4f}")
    print(f"Mean Recall: {mean_recall:.4f}")
    print(f"Mean F1-score: {mean_f1:.4f}")

    # Plot the confusion matrix
    cm = confusion_matrix(y_test, predicted_labels)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['B', 'M'], yticklabels=['B', 'M'])
    plt.title(f"Confusion Matrix")
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()

    # Plot the error curve for testing and validation
    plot_loss_curve(loss_curve)

    if NORMALIZATION_FLAG==1:
        # Model evaluation with cross-validation (when there is normalization)
        model = MLPClassifier(hidden_layer_sizes=NUM_HIDDEN_LAYERS, max_iter=MAX_ITER, random_state=RANDOM_STATE)
        cross_val_scores = cross_val_score(model, X_processed_scaled, y_processed, cv=num_folds)
        # Display cross-validation scores
        print(f"\nCross-validation Accuracy Scores: {cross_val_scores}")
        print(f"Mean Cross-validation Accuracy: {cross_val_scores.mean():.4f}")
        # Train the final model (when there is normalization)
        model.fit(X_processed_scaled, y_processed)
    else:
        # Create the MLP model (when there is no normalization)
        model = MLPClassifier(hidden_layer_sizes=NUM_HIDDEN_LAYERS, max_iter=MAX_ITER, random_state=RANDOM_STATE)
        # Train the model (when there is no normalization)
        model.fit(X_train, y_train)  # Removed: drop(['ID', 'Diagnosis'], axis=1)

    # Store the trained model and its parameters
    models_and_params.append({
        'model': model,
        'params': model.get_params()
    })

    return models_and_params

def get_mlp_parameters(params):
    # Get specific parameters of the MLP model
    parameters = []
    parameters.append(['hidden_layer_sizes', params['hidden_layer_sizes']])
    parameters.append(['activation', params['activation']])
    parameters.append(['solver', params['solver']])
    parameters.append(['alpha', params['alpha']])
    parameters.append(['max_iter', params['max_iter']])
    parameters.append(['random_state', params['random_state']])

    # Create a DataFrame with the parameters
    df_parameters = pd.DataFrame(parameters, columns=['Parameter', 'Value'])

    return df_parameters

def plot_learning_curve(model, X, y, num_folds=NUM_FOLDS):
    train_sizes, train_scores, test_scores = learning_curve(
        model, X, y, cv=num_folds, scoring='accuracy', train_sizes=np.linspace(0.1, 1.0, 10))

    plt.figure(figsize=(8, 5))
    plt.plot(train_sizes, np.mean(train_scores, axis=1), label='Training Score', color='navy')
    plt.plot(train_sizes, np.mean(test_scores, axis=1), label='Validation Score', color='red')
    plt.title('Learning Curve')
    plt.xlabel('Number of Training Examples')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()

def get_mlp_layers_info(model):
    # Get information about all hidden layers and the output layer of the MLP model
    layers_info = []

    # Add input layer (attributes)
    layers_info.append({
        'layer': 0,
        'num_neurons': model.coefs_[0].shape[0],  # Number of neurons in the input layer
        'bias': None  # There is no bias in the input layer
    })

    # Add hidden layers
    for i in range(len(model.coefs_)):
        # Number of neurons in hidden layer i
        num_neurons = model.coefs_[i].shape[1]

        # Bias of hidden layer i
        bias = model.intercepts_[i]

        layers_info.append({
            'layer': i + 1,
            'num_neurons': num_neurons,
            'bias': bias.tolist()  # Converting to list for easier display
        })

    # Add output layer
    layers_info.append({
        'layer': len(model.coefs_) + 1,
        'num_neurons': model.coefs_[-1].shape[0],  # Number of neurons in the output layer
        'bias': model.intercepts_[-1].tolist()  # Bias in the output layer
    })

    return layers_info


def plot_weights(model, feature_names):
    plt.figure(figsize=(10, 6))
    for i, coef in enumerate(model.coefs_):
        plt.subplot(1, len(model.coefs_), i + 1)
        plt.imshow(coef, interpolation='none', cmap='viridis')
        plt.title(f'Layer {i+1}')
        plt.xticks(range(len(feature_names)), feature_names, rotation=90)
        plt.yticks(range(coef.shape[0]))
        plt.colorbar()

    plt.tight_layout()
    plt.show()


def plot_roc_curve(model, X, y):
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=RANDOM_STATE)

    # Train the model
    model.fit(X_train, y_train)

    # Get the predicted probabilities for the positive class
    y_prob = model.predict_proba(X_test)[:, 1]

    # Calculate the ROC curve
    fpr, tpr, thresholds = roc_curve(y_test, y_prob)

    # Calculate the area under the ROC curve (AUC-ROC)
    roc_auc = auc(fpr, tpr)

    # Plot the ROC curve
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='blue', lw=2, label='ROC Curve (AUC = {:.2f})'.format(roc_auc))
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc='lower right')
    plt.show()

def plot_loss_curve(loss_curve):
    plt.figure(figsize=(8, 6))
    for fold, losses in enumerate(loss_curve, 1):
        plt.plot(losses, label=f'Fold {fold}')

    plt.title('Loss Curve Over Training')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

'''Initialization of the Main Code calling the Previous Function Blocks'''
if __name__ == "__main__":
    cancer_data = download_and_load_data()
    display_initial_table(cancer_data)
    display_data_summary(cancer_data)
    visualize_class_distribution(cancer_data)
    #visualize_column_distribution(cancer_data)

    # Preprocessing and imputing the data
    processed_data = preprocess_and_impute(cancer_data)
    X_processed = processed_data.drop(['ID', 'TARGET_COLUMN'], axis=1)
    y_processed = processed_data['TARGET_COLUMN']

    # Call the function and store the returned parameters
    model_params_all_folds = train_mlp_model(X_processed, y_processed)

    # Save the trained model to a file
    #import joblib
    #joblib.dump(model_params_all_folds, 'mlp_model.joblib')

    # Load the trained model from the file
    #loaded_model = joblib.load('mlp_model.joblib')

    # Display the parameters and layer information for all folds
    for fold, model_params in enumerate(model_params_all_folds, NUM_FOLDS):
        model = model_params['model']
        df_parameters = get_mlp_parameters(model_params['params'])
        layers_info = get_mlp_layers_info(model)

        # Display layer information in a table
        df_layers_info = pd.DataFrame(layers_info)
        print(f"\nModel Parameters:\n{df_parameters}")
        print(f"\nLayer Information:\n{df_layers_info}")
        print("NOTE: Remember that the hidden layers are those excluding the output and input layers.")

    # Call the function to plot the weights
    print("\nWeight Chart for Each of the Hidden Layers of the MLP Network")
    plot_weights(model, X_processed.columns)

    # Display the learning curve of the models
    print("\nChart of Learning Curves for Testing and Validation")
    plot_learning_curve(MLPClassifier(hidden_layer_sizes=NUM_HIDDEN_LAYERS, max_iter=MAX_ITER, random_state=RANDOM_STATE),
                X_processed, y_processed)

    # Create an instance of the MLPClassifier model
    mlp_model = MLPClassifier(hidden_layer_sizes=NUM_HIDDEN_LAYERS, max_iter=MAX_ITER, random_state=RANDOM_STATE)

    # Display ROC curve
    print("\nROC Curve Chart")
    plot_roc_curve(mlp_model, X_processed, y_processed)

```

CONSIDERATIONS:

If all metrics, including accuracy, precision, recall, and F1-score, reached 100%, this suggests that the model is performing perfectly on the specific dataset it was trained and evaluated on (even though this may be misleading due to Overfitting or Class Imbalance). However, based on the obtained metrics, we can make some analyses:

- Accuracy: All predictions made by the model are correct concerning the true label. Accuracy is the proportion of correct predictions to the total predictions.

- Precision: All instances that the model predicted as positive (class 'M') were indeed positive. Precision is the proportion of true positives to the total predicted as positive.

- Recall: All positive instances (class 'M') were correctly identified by the model. Recall is the proportion of true positives to the actual total of positives.

- F1-Score: The F1-Score is the harmonic mean of precision and recall. If both are 100%, the F1-Score will also be 100%.

Cross-Validation: If the model achieved 100% in all metrics across several folds during cross-validation, this is more promising and suggests good generalization.

In the previous code, in the training function module of the MLP network (train_mlp_model), to explicitly indicate the modification of parameters with normalization, the training of the network is written with indications of when Normalization is activated and when it is not activated.
